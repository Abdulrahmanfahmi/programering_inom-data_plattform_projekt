```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("../data/ledamoter.csv")


sns.set_theme(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='parti', order=df['parti'].value_counts().index, palette='viridis')
plt.title('Antal mandat per parti i Riksdagen')
plt.xlabel('Parti')
plt.ylabel('Antal ledamöter')
plt.show()
```
